```python
marksheet=[]
scorelist=[]
if __name__ == '__main__':
        for _ in range(int(input())):
                name = input()
                score = float(input())
                marksheet+=[[name,score]]
                scorelist+=[score]
        highScore=sorted(list(set(scorelist))) [1] #set to remove duplicates from array and converted to list
        #then sorted [1] pick second element

        print("marksheet =",marksheet) # marksheet = [['fdsj', 5.4], ['gjs', 7.4], ['ehwj', 7.2]]
        print("scorelist =",scorelist) # scorelist = [5.4, 7.4, 7.2]
        print("sortedScoreList =",highScore) # sortedScoreList = 7.2
        print(sorted(marksheet))
        for a,c in sorted(marksheet):
             if c==highScore:  
                print("Second least scored student",a)
```
